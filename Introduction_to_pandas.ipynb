If you want to add data to the dataframe, using a...
* series, the series doesn't have to match the length of the dataframe
* list, the length of the list has to match

```python
car_sales["Total fuel used (L)"] = car_sales["Odometer (KM)"]/100 * car_sales["Fuel per 100km"]
car_sales
```

```python
car_sales.drop("Total fuel used", axis=1, inplace=True)
```
